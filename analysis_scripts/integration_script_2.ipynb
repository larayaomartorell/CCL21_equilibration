```python
from scipy import integrate
import numpy as np
import matplotlib.pyplot as plt 
```

```python
with open('../data/time_averageF2.txt') as f:
    lines = f.readlines()
    time = [line.split()[0] for line in lines]
    force = [line.split()[1] for line in lines]
```

```python
print(time)
print(force)
```

```python
#linear velocity in A/ps
v2 = 0.05

#to initialize variables
integral = 0
i = 0
dt = 0

# elements in the list
N = len(force)

for i in range(N-1):
    dt = float(time[i+1]) - float(time[i])
    dx = v2*dt
    f = float(force[i])
    integral += f*dx 
    if float(time[i])*float(v2) > 13:
        break
    
print(integral)
```
